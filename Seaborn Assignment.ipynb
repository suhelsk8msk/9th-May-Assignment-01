Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans - 

Seaborn is a powerful Python visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1 Scatter Plot:
Use: Scatter plots are used to visualize the relationship between two variables. They are useful for identifying patterns, clusters, or correlations between the variables.

2 Histogram:
Use: Histograms are used to display the distribution of a single numerical variable. They help in understanding the underlying frequency distribution of the data.

3 Bar Plot:
Use: Bar plots are used to compare the values of different categories or groups. They are particularly useful for showing categorical data or comparing data across different groups.

4 Box Plot (Box-and-Whisker Plot):
Use: Box plots are used to visualize the distribution of a continuous variable across different categories or groups. They provide information about the central tendency, spread, and presence of outliers in the data.

5 Heatmap:
Use: Heatmaps are used to visualize the matrix-like data where each cell represents the value of the relationship between two variables. They are useful for identifying patterns and correlations in large datasets, especially in fields like genomics, finance, and social sciences.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. 
Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)
plt.title('Signal vs Timepoint for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event')
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Below is the code to load the "titanic" dataset using the load_dataset function of Seaborn and then plot two box plots:

Box plot of age distribution across different passenger classes ('pclass' on x-axis and 'age' on y-axis).

Box plot of fare distribution across different passenger classes ('pclass' on x-axis and 'fare' on y-axis).

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots
plt.figure(figsize=(12, 6))

# Box plot of age distribution across different passenger classes
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Age Distribution Across Passenger Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

# Box plot of fare distribution across different passenger classes
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Fare Distribution Across Passenger Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue for the 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)
plt.title('Histogram of Diamond Prices with Cut Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot with hue for the 'species' column
sns.pairplot(iris_data, hue="species")
plt.title('Pair Plot of Iris Dataset with Species')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Reshape the data for heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Passengers Heatmap Over Years and Months')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
