Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans - 

Matplotlib is a comprehensive library in Python used for creating static, animated, and interactive visualizations. It is widely used for data visualization due to its flexibility and ease of use. Matplotlib provides a wide range of plotting functions and customization options, making it suitable for various types of data visualization tasks.

Here are five common plots that can be created using the Pyplot module of Matplotlib:

1 - Line Plot: A basic plot where data points are connected by straight lines. It's useful for visualizing trends over continuous variables, such as time-series data.

2 - Scatter Plot: Scatter plots are used to display the relationship between two numerical variables. Each data point is represented by a marker on the plot, allowing visualization of patterns or correlations in the data.

3 - Histogram: Histograms are used to represent the distribution of a single numerical variable. It divides the data into bins and displays the frequency or density of data points within each bin.

4 - Bar Plot: Bar plots are used to compare categorical data. They represent the values of different categories as bars with heights proportional to the values they represent. Bar plots are useful for comparing quantities across different groups.

5 - Pie Chart: Pie charts are circular statistical graphics used to represent the proportional distribution of categorical data. Each slice of the pie represents a category, and the size of the slice corresponds to the proportion of data within that category.

Q2: What is a scatter plot? Use the following code to generate data for x and y. 

Using this generated data plot a scatter plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

Ans - 


A scatter plot is a type of plot used to visualize the relationship between two numerical variables. Each data point in the plot represents the value of one variable plotted along the x-axis and the value of another variable plotted along the y-axis. Scatter plots are useful for identifying patterns, trends, and correlations between variables.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
plt.scatter(x, y, color='blue', alpha=0.7)  # Alpha controls transparency
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)  # Add grid lines for better readability
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans -


The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to divide the figure into a grid of rows and columns and specify the position of each subplot within this grid. This is useful when you want to compare multiple plots side by side or create a complex layout of plots.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(12, 8))  # Adjust figure size as needed

# Subplot 1
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, position 1
plt.plot(x1, y1, color='blue')
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, position 2
plt.plot(x2, y2, color='green')
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, position 3
plt.plot(x3, y3, color='red')
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, position 4
plt.plot(x4, y4, color='purple')
plt.title('Line Plot 4')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Ans-

A bar plot is a type of graph that presents categorical data with rectangular bars. The lengths of the bars are proportional to the values they represent. Bar plots are used to compare different categories or groups of data and to visualize the distribution or frequency of categorical variables.

Bar plots are useful for the following purposes:

1- Comparison: They allow easy visual comparison of the values of different categories.

2- Trend Analysis: They help in identifying trends or patterns within categorical data.

3- Highlighting Differences: They can effectively highlight differences or disparities between categories.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.grid(axis='y')  # Add gridlines only on the y-axis
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')
plt.title('Company Profits')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.grid(axis='x')  # Add gridlines only on the x-axis
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Ans-

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. It provides a visual summary of the central tendency, variability, and skewness of the data along with potential outliers.

Box plots are used for the following purposes:

1- Identifying Central Tendency: The line inside the box represents the median value of the dataset, providing information about the central tendency.

2- Assessing Spread: The length of the box indicates the interquartile range (IQR), which gives a measure of the spread or variability of the data.

3- Detecting Skewness: The shape of the box and whiskers can give insights into the skewness of the data distribution.

4- Spotting Outliers: Any data points lying beyond the whiskers are considered potential outliers, helping in identifying unusual observations in the dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine data into a list
data = [box1, box2]

# Create box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
